<h3>Example 1: Matplotlib</h3>

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(33 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='Pretty fun, no?')
ax.grid()

fig.savefig("test.png")
plt.show()
```

```python
# adding Pandas - example of boxplot:
# Boxplot
# but some concerns - the total cannot be removed (which is affecting
# the size of the other data; we don't need state and legendary columns 
# because they don't need them)
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('vacation-utf8.csv', index_col=0)
df.head()

#scatter plot: using a regression
# recommended way
sns.lmplot(x='rank1', y='rank2', data=df)
sns.boxplot(data=df)

# Pre-format DataFrame
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
 
# New boxplot using stats_df
sns.boxplot(data=stats_df)
```

```python
## finally make it look prettier?
## seaborn has other themes.  Default is "darkgrid"; change to 
# whitegrid and change the visualization technique to "violin".
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('vacation-utf8.csv', index_col=0)
df.head()

#scatter plot: using a regression
# recommended way
sns.lmplot(x='rank1', y='rank2', data=df)

# Set theme
sns.set_style('whitegrid')
 
# Violin plot
sns.violinplot(x='type', y='rank1', data=df)

# define colors:
pkmn_type_colors = ['#78C850',  # forest (forêt)
                    '#F08030',  # woods (bois)
                    '#6890F0',  # park (parc)
                    '#A8B820',  # beach (plage)
                    '#A8A878',  # domaine 
                    '#A040A0',  # camping
                   ]
# Violin plot with vacation color palette
sns.violinplot(x='type', y='rank1', data=df, 
               palette=pkmn_type_colors) # Set color palette

# swarm plot
# Swarm plot with vacation color palette
sns.swarmplot(x='type', y='rank1', data=df, 
              palette=pkmn_type_colors)
```

```python
## overlay plots to make more data standout...
# Set figure size with matplotlib
# define colors:
pkmn_type_colors = ['#78C850',  # forest (forêt)
                    '#F08030',  # woods (bois)
                    '#6890F0',  # park (parc)
                    '#A8B820',  # beach (plage)
                    '#A8A878',  # domaine 
                    '#A040A0',  # camping
                   ]
plt.figure(figsize=(10,6))
 
# Create plot
sns.violinplot(x='type',
               y='rank1', 
               data=df, 
               inner=None, # Remove the bars inside the violins
               palette=pkmn_type_colors)
 
sns.swarmplot(x='type', 
              y='rank1', 
              data=df, 
              color='k', # Make points black
              alpha=0.7) # and slightly transparent
```

```python
""" As you can see, all of our stats are in separate columns. Instead, we want to "melt" them into one column.

To do so, we'll use Pandas's melt() function. It takes 3 arguments:

First, the DataFrame to melt.
Second, ID variables to keep (Pandas will melt all of the other ones).
Finally, a name for the new, melted variable.
Here's the output: """
stats_df.head()

# Melt DataFrame
# Melt DataFrame
melted_df = pd.melt(stats_df, 
                    id_vars=["Name", "type", "town"], # Variables to keep
                    var_name="Stat") # Name of melted variable
melted_df.head()
```

```python
""" All 6 of the stat columns have been "melted" into one, and the new Stat column indicates the original stat (HP, Attack, Defense, Sp. Attack, Sp. Defense, or Speed). For example, it's hard to see here, but Bulbasaur now has 6 rows of data.

In fact, if you print the shape of these two DataFrames...

Shape Comparison """
print( stats_df.shape )
print( melted_df.shape )
# (151, 9)
# (906, 5)

""" SwarmPlot with melted_df - change the hue to color points by Pokumon types """
# Swarmplot with melted_df
sns.swarmplot(x='Stat', y='value', data=melted_df, 
              hue='type')

```

```python
""" Heatmap """
# Calculate correlations
corr = stats_df.corr()
 
# Heatmap
sns.heatmap(corr)

```

```python

""" Joint Distribution Plot """
# Joint Distribution Plot
sns.jointplot(x='rank1', y='rank2', data=df)
```
